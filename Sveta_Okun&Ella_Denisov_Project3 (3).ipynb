# Project3 - Superstore Sales Data Analysis

#### Project Goals:
The goal of this project is to analyze retail sales data from a Superstore dataset in order to understand patterns in sales, profitability, and customer segments.

This project analyzes the Superstore sales dataset in order to understand how different factors influence business performance. The goal is to understand how these factors relate to business performance and to identify patterns that can help decision makers make better strategic decisions and improve their marketing and sales strategies.

Dataset: Superstore Dataset  
https://www.kaggle.com/datasets/vivek468/superstore-dataset-final

#### Research Questions
1. Which product categories & sub-categories generate the highest sales & profit?
2. Which states perform best in terms of sales & profit?
3. How do sales change over time?
4. Which segment performs best in sales?
5. How discount impacts profit?
6. What relationship exists between sales, profit, quantity and discount?

### Data Cleaning

```python
df.isnull().sum()
```

```python
# Converting date columns to datetime:
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Creatimg additional time columns for analysis:
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month

df.head()
```

### Descriptive Statistical Analysis

```python
# Sales by category:
sales_by_category = df.groupby('Category')['Sales'].sum().reset_index()
sales_by_category
```

### Advanced Analysis

```python
# Sales and Profit by Category and Region:
sales_profit_by_category_region = df.groupby(['Category', 'Region'])[['Sales', 'Profit']].agg(['mean', 'min', 'max','sum', 'count']).reset_index()
sales_profit_by_category_region
```

```python
# Top 10 Most Profitable Products by Sub-Category:
top_profit_subcategory = df.groupby('Sub-Category')['Profit'].agg('sum').sort_values(ascending=False).head(10).to_frame()
top_profit_subcategory
```

```python
# States with Negative Profit:
loss_states = (df.groupby('State')['Profit'].sum().to_frame())
loss_states = loss_states[loss_states['Profit'] < 0].sort_values(by='Profit')
loss_states
```

```python
# Impact of Discounts on Profit:
discount_profit = df.groupby('Discount')['Profit'].mean().to_frame()
discount_profit
```

```python
# Monthly Sales Trend by Year:
monthly_sales = (df.groupby(['Year','Month']).agg(total_sales=('Sales','sum'),avg_sales=('Sales','mean'),
                order_count=('Order ID','nunique')).reset_index())
monthly_sales
```

### Data Visualization

```python
# Total Sales and Profit by Product Category - bar chart:
category_analysis = (df.groupby('Category').agg(total_sales=('Sales','sum'),total_profit=('Profit','sum'))
                     .sort_values(by='total_sales'))

category_analysis.plot(figsize=(10,6), kind='bar', color=['darkkhaki','teal'])
plt.title('Total Sales and Profit by Product Category', fontsize=16, color='black', fontfamily='monospace')
plt.style.use('ggplot')
plt.xlabel('Category', fontsize=12, color='black', fontfamily='monospace')
plt.ylabel('Amount', fontsize=12, color='black', fontfamily='monospace')
plt.xticks(rotation=0)
plt.legend(loc = 'upper left')
plt.show()
```

This plot shows that not all high-sales categories generate proportional profit. "Technology" stands out as the most profitable category, "Office Supplies" provides steady but moderate performance and "Furniture" appears to be less efficient in converting sales into profit. These insights could help the businesse focus more on high-performing categories while reviewing pricing or cost strategies for lower-profit categories.

```python
# Total Sales and Profit by Product Sub-Category - bar chart:
sub_category_analysis = (df.groupby('Sub-Category').agg(total_sales=('Sales','sum'),total_profit=('Profit','sum'))
                     .sort_values(by='total_sales'))

sub_category_analysis.plot(figsize=(18,6), kind='bar', color=['darkkhaki','teal'])
plt.title('Total Sales and Profit by Product Sub-Category', fontsize=16, color='black', fontfamily='monospace')
plt.style.use('ggplot')
plt.xlabel('Sub-Category', fontsize=12, color='black', fontfamily='monospace')
plt.ylabel('Amount', fontsize=12, color='black', fontfamily='monospace')
plt.xticks(rotation=0)
plt.legend(loc = 'upper left')
plt.show()
```

The bar chart presents total sales and total profit across different product sub-categories, providing a detailed view of product-level performance.
Product performance should be evaluated not only by sales volume but also by profitability. Based on this analysis, some sub-categories may exhibit low or even negative profit despite generating sales, highlighting potential issues such as over-discounting, high costs, or inefficient pricing strategies.

```python
# Top 10 States by Sales vs Profit - bar chart:
state_analysis = (df.groupby('State').agg(total_sales=('Sales','sum'),total_profit=('Profit','sum'))
                .sort_values(by='total_sales', ascending=False).head(10))

state_analysis = state_analysis.sort_values(by='total_sales')

ax = state_analysis.plot(figsize=(10,8), kind='barh', color=['darkkhaki','teal'])
plt.title('Top 10 States by Sales vs Profit', fontsize=16, color='black', fontfamily='monospace')
plt.xlabel('Sales & Profit', fontsize=12, color='black', fontfamily='monospace')
plt.ylabel('State', fontsize=12, color='black', fontfamily='monospace')
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', padding=3, fontsize=8)
plt.legend(title='Metric')
plt.show()
```

This plot presents the top 10 states based on total sales and compares their profit levels. While some states such as California and New York demonstrate strong performance in both sales and profit, other states show a gap between sales and profitability. This comparison helps identify the most valuable geographic markets and highlights areas where profitability can be improved.

```python
# Monthly Sales Trend by Year - Line plot:
monthly_sales_unstack = (df.groupby(['Year','Month'])['Sales'].sum().unstack(level=0))
monthly_sales_unstack.index = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

monthly_sales_unstack.plot(figsize=(10,5), kind='line', marker='o', color=['cadetblue','rosybrown','darkkhaki','olivedrab'])
plt.title('Monthly Sales Trend by Year', fontsize=16, color='black', fontfamily='monospace')
plt.xlabel('Month', fontsize=12, color='black', fontfamily='monospace')
plt.ylabel('Total Sales', fontsize=12, color='black', fontfamily='monospace')
plt.xticks(ticks=range(12), labels=monthly_sales_unstack.index, rotation=0)
plt.legend(title='Year')
plt.grid(True)
plt.show()
```

This chart presents the monthly sales trends across different years, comparison of seasonal patterns and performance over time. Overall, sales tend to fluctuate throughout the year, showing periods of both growth and decline. We can conclude that sales generally increase toward the end of the year, particularly in the final months, suggesting strong seasonal demand. In contrast, earlier months tend to show lower sales levels.

```python
# Sales Distribution by Customer Segment - Pie chart:
segment_sales = df.groupby('Segment')['Sales'].sum()

colors = ['cadetblue','rosybrown','darkkhaki']

plt.figure(figsize=(7,7))
segment_sales.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Sales Distribution by Customer Segment',fontsize=16, color='black', fontfamily='monospace')
plt.ylabel('')
plt.show()
```

The Pie Chart shows that the Consumer segment holds the largest share of sales, indicating that individual customers are the primary drivers of revenue. Corporate customers contribute a significant portion, while Home offices represent a smaller segment.

```python
# Discount vs Profit - Scatter plot:
plt.figure(figsize=(10,6))  
plt.scatter(df['Discount'] * 100, df['Profit'], color='teal', alpha=0.6, s=40)
plt.title('Discount vs Profit Relationship', fontsize=16, color='black', fontfamily='monospace')
plt.xlabel('Discount (%)')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```

This plot shows the relationship between discount levels and profit for individual transactions. 
Each point represents a sale, from this we can conclude that higher discount levels are generally associated with lower profit, and in many cases, transactions with high discounts result in negative profit. This indicates that excessive discounting can significantly reduce profitability and may even lead to losses.

```python
# Correlation Matrix - Heatmap:
sales_corr = ['Sales','Profit','Quantity','Discount']

corr_matrix = df[sales_corr].corr()
plt.figure(figsize=(8,4))
sns.heatmap(corr_matrix, annot=True, cmap='BrBG')
plt.title('Correlation Matrix', fontsize=16, color='black', fontfamily='monospace')
plt.show()
```

The HeatMap shows:
1. Sales & Profit- weak to moderate positive correlation: the company generates revenue effectively, but profitability is not always consistent.
2. Discount & Profit- negative correlation: discount strategy has a direct negative impact on profitability, and high discounting may harm the business.
3. Quantity & Sales- positive correlation: selling more units leads to higher total sales.
4. Quantity & Profit- Weak correlation: selling more units does not always increase profit significantly.

### Conclusions

- The analysis reveals that business performance cannot be evaluated based on sales alone, as high revenue does not necessarily translate into high profitability. This is evident across product categories, sub-categories, and geographic regions, where some areas generate strong sales but relatively low or even negative profit. 
- A key finding is that discount strategy has a significant impact on profitability. Both the scatter plot and heatmap show a clear negative relationship between discount and profit, indicating that excessive discounting often leads to reduced margins and, in many cases, financial losses. This highlights the importance of maintaining a balanced pricing strategy.
- At the product level, certain sub-categories (such as Phones and Chairs) dominate sales, but not all of them contribute equally to profit. Some sub-categories generate low or negative profit despite strong sales, suggesting issues such as high costs or inefficient pricing. This emphasizes the need to focus on high-margin products rather than just high-volume sales.
- From the analysis, we conclude that sales generally increase toward the end of the year, particularly in the final months, suggesting strong seasonal demand. 
In contrast, earlier months tend to show lower sales levels. This highlights the importance of time-based analysis for planning and forecasting.
- In terms of customer behavior, the Consumer segment is the primary driver of sales, indicating strong dependence on individual customers. 
While Corporate customers also contribute significantly, there is an opportunity to further expand in this segment.
- Geographically, the analysis shows that some states are more efficient than others in converting sales into profit, reinforcing the need for targeted state strategies.

- Overall, the results show that the business needs to not only increase sales, but also make sure those sales are profitable. It is important to manage discounts carefully, focus on products and customer groups that perform well, and use insights from seasonal trends and different states to make better business decisions.
