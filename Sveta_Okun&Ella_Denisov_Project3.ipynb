# Project3 - Superstore Sales Data Analysis

#### Project Goals:
The goal of this project is to analyze retail sales data from a Superstore dataset in order to understand patterns in sales, profitability, and customer segments.

This project analyzes the Superstore sales dataset in order to understand how different factors influence business performance. The goal is to understand how these factors relate to business performance and to identify patterns that can help decision makers make better strategic decisions and improve their marketing and sales strategies.

Dataset: Superstore Dataset  
https://www.kaggle.com/datasets/vivek468/superstore-dataset-final

#### Research Questions
1. Which product categories generate the highest sales and profit?
2. Which regions perform best in terms of revenue?
3. How do sales change over time?
4. What relationship exists between sales and profit?
5. Which customer segments contribute the most to revenue?

Tools and Techniques: Python, Pandas, GroupBy, aggregation, Data visualization

### Column Description

- Row ID- Unique ID for each row 
- Order ID- Unique Order ID for each Customer  
- Order Date- Date when the order was placed  
- Ship Date- Date when the order was shipped  
- Ship Mode- Shipping Mode specified by the Customer 
- Customer ID- Unique ID to identify each Customer 
- Customer Name- Customer name  
- Segment- Customer segment (Consumer, Corporate, Home Office)  
- Country/City/State/Region- Geographic information  
- Product ID- Unique ID of the Product  
- Category- Product category  
- Sub Category- Product sub category  
- Product Name- Name of the Product
- Sales- Sales of the Product  
- Quantity- Number of items sold  
- Discount- Discount provided 
- Profit- Profit/Loss from the order


### Data Cleaning

```python
df.isnull().sum()
```

```python
# Converting date columns to datetime:
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Creatimg additional time columns for analysis:
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month

df.head()
```

### Descriptive Statistical Analysis

```python
# Identifying extreme values:
df[['Sales','Profit']].agg(['min','max'])
```

```python
# how many categories are there in the superstore dataset?
df["Category"].nunique()
```

```python
df["Category"].unique()
```

```python
# Sales by category:
sales_by_category = df.groupby('Category')['Sales'].sum().reset_index()
sales_by_category
```

```python
# Descriptive statistics of sales by category:
df.groupby('Category')['Sales'].describe()
```

```python
# Which product categories are the most profitable:
profit_by_category = df.groupby('Category')['Profit'].sum().reset_index()
profit_by_category
```

```python
df.groupby('Category')['Profit'].describe()
```

```python
# Customer segments represent different types of customers:
df.groupby('Segment')['Sales'].describe()
```

```python
# Sales by Customer Segment:
customer_segments = df.groupby('Segment')['Sales'].sum().round(2).reset_index()
customer_segments
```

```python
# Top 10 states by sales:
top_states_sales = (df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)).round(1).to_frame()
top_states_sales
```

```python
# Top 10 states by profit:
top_states_profit = (df.groupby('State')['Profit'].sum().sort_values(ascending=False).head(10)).round(1).to_frame()
top_states_profit
```

```python
# Avarage discount by category:
avg_discount_by_category = df.groupby('Category')['Discount'].mean().to_frame()
avg_discount_by_category
```

```python
# Quantity of items sold by category:
df.groupby('Category')['Quantity'].count().to_frame()
```

```python
# Sales by sub-category:
df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).to_frame()
```

```python
# Total sales by category and sub-category:
sales_by_category_subcategory = df.groupby(["Category", "Sub-Category"])["Sales"].sum().reset_index()
sales_by_category_subcategory
```

### Advanced Analysis

```python
# Sales and Profit by Category and Region:
sales_profit_by_category_region = df.groupby(['Category', 'Region'])[['Sales', 'Profit']].agg(['mean', 'min', 'max','sum', 'count']).reset_index()
sales_profit_by_category_region
```

```python
# Top 10 Most Profitable Products by Sub-Category:
top_profit_subcategory = df.groupby('Sub-Category')['Profit'].agg('sum').sort_values(ascending=False).head(10).to_frame()
top_profit_subcategory
```

```python
# States with Negative Profit:
loss_states = (df.groupby('State')['Profit'].sum().to_frame())
loss_states = loss_states[loss_states['Profit'] < 0].sort_values(by='Profit')
loss_states
```

```python
# Impact of Discounts on Profit:
discount_profit = df.groupby('Discount')['Profit'].mean().to_frame()
discount_profit
```

```python
# Monthly Sales Trend by Year:
monthly_sales = (df.groupby(['Year','Month']).agg(total_sales=('Sales','sum'),avg_sales=('Sales','mean'),
                order_count=('Order ID','nunique')).reset_index())
monthly_sales
```

### Data Visualization

```python
# Total Sales and Profit by Product Category - bar chart:
category_analysis = (df.groupby('Category').agg(total_sales=('Sales','sum'),total_profit=('Profit','sum'))
                     .sort_values(by='total_sales'))

category_analysis.plot(figsize=(10,6), kind='bar', color=['darkkhaki','teal'])
plt.title('Total Sales and Profit by Product Category', fontsize=16, color='black', fontfamily='monospace')
plt.style.use('ggplot')
plt.xlabel('Category', fontsize=12, color='black', fontfamily='monospace')
plt.ylabel('Amount', fontsize=12, color='black', fontfamily='monospace')
plt.xticks(rotation=0)
plt.legend(loc = 'upper left')
plt.show()
```

This plot shows that not all high-sales categories generate proportional profit. "Technology" stands out as the most profitable category, "Office Supplies" provides steady but moderate performance and "Furniture" appears to be less efficient in converting sales into profit.

```python
# Top 10 States by Sales vs Profit - bar chart:
state_analysis = (df.groupby('State').agg(total_sales=('Sales','sum'),total_profit=('Profit','sum'))
                .sort_values(by='total_sales', ascending=False).head(10))

state_analysis = state_analysis.sort_values(by='total_sales')

ax = state_analysis.plot(figsize=(10,8), kind='barh', color=['darkkhaki','teal'])
plt.title('Top 10 States by Sales vs Profit', fontsize=16, color='black', fontfamily='monospace')
plt.xlabel('Sales & Profit', fontsize=12, color='black', fontfamily='monospace')
plt.ylabel('State', fontsize=12, color='black', fontfamily='monospace')
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', padding=3, fontsize=8)
plt.legend(title='Metric')
plt.show()
```

This plot presents the top 10 states based on total sales and compares their profit levels. While some states such as California and New York demonstrate strong performance in both sales and profit, other states show a gap between sales and profitability.

```python
# Monthly Sales Trend by Year - Line plot:
monthly_sales_unstack = (df.groupby(['Year','Month'])['Sales'].sum().unstack(level=0))
monthly_sales_unstack.index = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

monthly_sales_unstack.plot(figsize=(10,5), kind='line', marker='o', color=['cadetblue','rosybrown','darkkhaki','lightcoral'])
plt.title('Monthly Sales Trend by Year', fontsize=16, color='black', fontfamily='monospace')
plt.xlabel('Month', fontsize=12, color='black', fontfamily='monospace')
plt.ylabel('Total Sales', fontsize=12, color='black', fontfamily='monospace')
plt.xticks(ticks=range(12), labels=monthly_sales_unstack.index, rotation=0)
plt.legend(title='Year')
plt.grid(True)
plt.show()
```

This chart presents the monthly sales trends across different years, comparison of seasonal patterns and performance over time. Overall, sales tend to fluctuate throughout the year, showing periods of both growth and decline.

```python
# Profit Distribution by Sub-Category - Pie Chart:
subcat_sales = (df.groupby('Sub-Category')['Sales'].agg('sum').sort_values(ascending=False))

subcat_sales.plot(figsize=(18,16), kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution by Sub-Category', fontsize=16, color='black', fontfamily='monospace')
plt.ylabel('') 
plt.show()
```

The pie chart shows the distribution of total sales across different product sub-categories.
Overall, this plot helps to quickly understand the relative importance of each sub-category and supports better decision-making regarding product focus and business strategy.

```python
# Discount vs Profit - Scatter plot:
plt.figure(figsize=(10,6))  
plt.scatter(df['Discount'] * 100, df['Profit'], color='teal', alpha=0.6, s=40)
plt.title('Discount vs Profit Relationship', fontsize=16, color='black', fontfamily='monospace')
plt.xlabel('Discount (%)')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```

This plot shows the relationship between discount levels and profit for individual transactions. 
Each point represents a sale, from this we can conclude how profitability changes as discounts increase.

```python
plt.figure(figsize=(10,6))

plt.hist(
    df['Profit'],
    bins=30,
    color='teal',
    edgecolor='black'
)

plt.title('Profit Distribution', fontsize=16)
plt.xlabel('Profit')
plt.ylabel('Frequency')

plt.grid(True)

plt.show()
```
